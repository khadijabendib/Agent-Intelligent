```python
import random
import numpy as np
def init_matrix(n, m, prob: float):
    matrix = np.random.choice([0, 1], size=(n * m), p=[prob, 1 - prob])
    return np.reshape(matrix, (n, m))



def Actualiser_etat(matrix):
    coord_list=[]
    result = []
    top, bottom, left, right = 0, len(matrix)-1, 0, len(matrix[0])-1
    direction="right"
    while top <= bottom and left <= right:
        if direction == "right":
            for i in range(left, right+1):
                aspirer(matrix,top,i)
                if matrix[top][i] == 1:
                    coord_list.append((top, i))
                result.append("right")
            top += 1
            direction = "bottom"
        elif direction == "bottom":
            for i in range(top, bottom+1):
                aspirer(matrix,i,right)
                if matrix[i][right] == 1:
                    coord_list.append((i, right))
                result.append("bottom")
            right -= 1
            direction = "left"
        elif direction == "left":
            for i in range(right, left-1, -1):
                aspirer(matrix,bottom,i)
                if matrix[bottom][i] == 1:
                    coord_list.append((bottom, i))
                result.append("left")
            bottom -= 1
            direction = "top"
        elif direction == "top":
            for i in range(bottom, top-1, -1):
                aspirer(matrix,i,left)
                if matrix[i][left] == 1:
                    coord_list.append((i, left))
                result.append("top")
            left += 1
            direction = "right"
    return result,coord_list


def Actualiser_etat_4(matrix):
    coord_list=[]
    result = []
    row, col = 0, 0
    direction = "right"
    while True:
        aspirer(matrix, row, col)
        if matrix[row][col] == 1:
            coord_list.append((row, col))
        result.append((row, col))
        if row == len(matrix)-1 and col == len(matrix[0])-1:
            break
        # Vérifie la case suivante dans la direction actuelle
        next_row, next_col = row, col
        if direction == "right":
            next_col += 1
        elif direction == "left":
            next_col -= 1
        elif direction == "up":
            next_row -= 1
        elif direction == "down":
            next_row += 1
        # Si la case suivante n'est pas accessible ou si c'est la dernière case, change de direction
        if (next_row < 0 or next_row >= len(matrix) or next_col < 0 or next_col >= len(matrix[0])):
            if direction == "right":
                direction = "down"
            elif direction == "down":
                direction = "left"
            elif direction == "left":
                direction = "up"
            elif direction == "up":
                direction = "right"
        # Déplace l'aspirateur dans la nouvelle direction
        if direction == "right":
            col += 1
        elif direction == "left":
            col -= 1
        elif direction == "up":
            row -= 1
        elif direction == "down":
            row += 1
    return result, coord_list

def Actualiser_etat_et_glissement(matrix):
    coord_list=[]
    result = []
    top, bottom, left, right = 0, len(matrix)-1, 0, len(matrix[0])-1
    direction="right"
    while top <= bottom and left <= right:
        if direction == "right":
            for i in range(left, right+1):
                aspirer(matrix,top,i)
                if matrix[top][i] == 1:
                    coord_list.append((top, i))
                result.append("right")
            top += 1
            direction = "bottom"
        elif direction == "bottom":
            for i in range(top, bottom+1):
                aspirer(matrix,i,right)
                if matrix[i][right] == 1:
                    coord_list.append((i, right))
                result.append("bottom")
            right -= 1
            direction = "left"
        elif direction == "left":
            for i in range(right, left-1, -1):
                aspirer(matrix,bottom,i)
                if matrix[bottom][i] == 1:
                    coord_list.append((bottom, i))
                result.append("left")
            bottom -= 1
            direction = "top"
        elif direction == "top":
            for i in range(bottom, top-1, -1):
                aspirer(matrix,i,left)
                if matrix[i][left] == 1:
                    coord_list.append((i, left))
                result.append("top")
            left += 1
            direction = "right"
            # Ajouter une condition pour simuler les roues qui glissent avec une probabilité de 10%
        if random.random() < 0.1:
                result.pop()
    return result,coord_list

def aspirer(x, m, n):
    if x[m][n] == 1:
        rand=random.random()
        # print(rand)
        if rand < 0.2:
            print("L'action a échoué")
        else:
            x[m][n] = 0  # Changer la valeur de la case sale en 0 pour l'indiquer comme nettoyée
    else:
        print("La case est déjà propre")
    return x

def salissage(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i][j] == 0 and random.random() < 0.05:
                matrix[i][j] = 1


def règle1(list,left,right,top):
    for i in range(left, right + 1):
        list.append("right")
    top += 1
    direction = "bottom"
    return list,direction

def règle2(list,right,bottom,top):
    for i in range(top, bottom + 1):
        list.append("bottom")
    right -= 1
    direction = "left"
    return list,direction

def règle3(list,left,right,bottom):
    for i in range(right, left - 1, -1):
        list.append("left")
    bottom -= 1
    direction = "top"
    return list,direction

def règle4(list,left,top,bottom):
    for i in range(bottom, top - 1, -1):
        list.append("top")
    left += 1
    direction = "right"
    return list,direction


def trouver_règle(list,direction,left,right,top,bottom):
    if direction=="right":
        list,direction=règle1(list,left,right,top)
    if direction=="bottom":
        list,direction=règle2(list,right,bottom,top)
    if direction == "left":
        list,direction,=règle3(list,left,right,bottom)
    if direction == "top":
       list,direction=règle4(list, left, right, bottom)


if __name__ == '__main__':
    x=int(input("Entrer le nombre de ligne de la matrice :"))
    y=int(input("Entrer le nombre de la colonne de la matrice:"))
    matrix=init_matrix(x,y,0.7)
    matrix1=matrix.copy()
    print(matrix)
    Actualiser_etat_et_glissement(matrix)
    print("Apres le nettoyage with fonction d'actualiser etat et glissement :")
    print(matrix)
    print("Apres l'appel de fonction de salissage:")
    salissage(matrix)
    print(matrix)
    print("C'est ça où l'agent sait pas l'environnement:")
    Actualiser_etat_4(matrix1)
    print(matrix1)


```
